```python
# -*- coding: utf-8 -*-
"""
Task2.ipynb
"""

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

"""# 5-Means Clustering"""

# reading data into Data Frame
data_frame = pd.read_csv('Task 2 Input.csv')

# selecting features Open and Adj Close
X = data_frame[['Open', 'Adj Close']]

# number of clusters
k=5

# creating a k-means cluster
kmeans = KMeans(n_clusters=k, init='random', n_init=100).fit(X)

# initializing new data frame to hold cluster classification
data_frame = pd.DataFrame(data = X, columns=['Open', 'Adj Close'])

# appending Cluster column with cluster labels for each row
data_frame['Cluster'] = kmeans.labels_

# extracting data for each cluster
cluster_1 = data_frame[data_frame['Cluster'] == 0]
cluster_2 = data_frame[data_frame['Cluster'] == 1]
cluster_3 = data_frame[data_frame['Cluster'] == 2]
cluster_4 = data_frame[data_frame['Cluster'] == 3]
cluster_5 = data_frame[data_frame['Cluster'] == 4]

# plotting clusters
plt.scatter(cluster_1['Open'], cluster_1['Adj Close'], c='red')
plt.scatter(cluster_2['Open'], cluster_2['Adj Close'], c='blue')
plt.scatter(cluster_3['Open'], cluster_3['Adj Close'], c='green')
plt.scatter(cluster_4['Open'], cluster_4['Adj Close'], c='yellow')
plt.scatter(cluster_5['Open'], cluster_5['Adj Close'], c='orange')

# plotting cluster centers
cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', s=50)

# adding graph labels
plt.title('{}-means clustering'.format(k))
plt.xlabel('Open')
plt.ylabel('Adj Close')

# displaying plot graph
plt.show()

# displaying SSE
print('SSE = {}'.format(kmeans.inertia_))

"""# 7-Means Clustering"""

# reading data into Data Frame
data_frame = pd.read_csv('Task 2 Input.csv')

# selecting features Open and Adj Close
X = data_frame[['Open', 'Adj Close']]

# number of clusters
k=7

# creating a k-means cluster
kmeans = KMeans(n_clusters=k, init='random', n_init=100).fit(X)

# initializing new data frame to hold cluster classification
data_frame = pd.DataFrame(data = X, columns=['Open', 'Adj Close'])

# appending Cluster column with cluster labels for each row
data_frame['Cluster'] = kmeans.labels_

# extracting data for each cluster
cluster_1 = data_frame[data_frame['Cluster'] == 0]
cluster_2 = data_frame[data_frame['Cluster'] == 1]
cluster_3 = data_frame[data_frame['Cluster'] == 2]
cluster_4 = data_frame[data_frame['Cluster'] == 3]
cluster_5 = data_frame[data_frame['Cluster'] == 4]
cluster_6 = data_frame[data_frame['Cluster'] == 6]
cluster_7 = data_frame[data_frame['Cluster'] == 7]

# plotting clusters
plt.scatter(cluster_1['Open'], cluster_1['Adj Close'], c='red')
plt.scatter(cluster_2['Open'], cluster_2['Adj Close'], c='blue')
plt.scatter(cluster_3['Open'], cluster_3['Adj Close'], c='green')
plt.scatter(cluster_4['Open'], cluster_4['Adj Close'], c='yellow')
plt.scatter(cluster_5['Open'], cluster_5['Adj Close'], c='orange')
plt.scatter(cluster_6['Open'], cluster_6['Adj Close'], c='purple')
plt.scatter(cluster_7['Open'], cluster_7['Adj Close'], c='grey')

# plotting cluster centers
cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', s=50)

# adding titles to graph
plt.title('{}-means clustering'.format(k))
plt.xlabel('Open')
plt.ylabel('Adj Close')

# displaying plot graph
plt.show()

# diplaying SSE
print('SSE = {}'.format(kmeans.inertia_))

"""# 2-Means Clustering"""

# reading data into Data Frame
data_frame = pd.read_csv('Task 2 Input.csv')

# selecting features Open and Adj Close
X = data_frame[['Open', 'Adj Close']]

# number of clusters
k=2

# creating a k-means cluster
kmeans = KMeans(n_clusters=k, init='random', n_init=100).fit(X)

# initializing new data frame to hold cluster classification
data_frame = pd.DataFrame(data = X, columns=['Open', 'Adj Close'])

# appending Cluster column with cluster labels for each row
data_frame['Cluster'] = kmeans.labels_

# extracting data for each cluster
cluster_1 = data_frame[data_frame['Cluster'] == 0]
cluster_2 = data_frame[data_frame['Cluster'] == 1]

# plotting clusters
plt.scatter(cluster_1['Open'], cluster_1['Adj Close'], c='red')
plt.scatter(cluster_2['Open'], cluster_2['Adj Close'], c='blue')

# plotting cluster centers
cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', s=50)

# adding titles to graph
plt.title('{}-means clustering'.format(k))
plt.xlabel('Open')
plt.ylabel('Adj Close')

# displaying plot graph
plt.show()

# diplaying SSE
print('SSE = {}'.format(kmeans.inertia_))

#
new_df = pd.read_csv('task6.csv')
new_df.drop(['Volume', 'Company', 'Direction'], axis=1, inplace=True)

new_df['Direction'] = (new_df['Adj Close'] - new_df['Open'] >= 0).astype(int)

new_df['Cluster'] = data_frame['Cluster']

print(new_df)

matches = len(new_df[new_df['Direction'] == new_df['Cluster']])
total = len(data_frame)

print('Accuracy: {}'.format(matches / total * 100))
```
