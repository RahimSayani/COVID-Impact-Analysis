```python
# -*- coding: utf-8 -*-
"""
Task4.ipynb
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# creating data frame containing the data set
data_frame = pd.read_csv('217005364-T0.csv')

# extracting features of the data set
features = data_frame[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]

# standardizing features
std_features = StandardScaler().fit_transform(features)

# reducing the features into principal components
pca = PCA()
principal_components = pca.fit_transform(std_features)

# comparing principal components
evr = pca.explained_variance_ratio_

for pc_evr in evr:
  print("Principal Component EVR: ", pc_evr)

# creating a data frame with the first 2 principal components
pc_data_frame = pd.DataFrame(data=principal_components[:,:2], columns=['PC1', 'PC2'])

# printing the table of principal components
print("\n", pc_data_frame)

# visualizing PC1 and PC2 in a two component graph
pc1 = principal_components[:, 0]
pc2 = principal_components[:, 1]

# initializing scatter plot
plot.scatter(pc1, pc2)

# adjusting scatter plot labels
plot.xlabel("Principal Component 1")
plot.ylabel("Principal Component 2")
plot.title("2 Component PCA")

# display scatter plot
plot.show()
```
