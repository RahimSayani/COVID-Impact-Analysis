```python
# -*- coding: utf-8 -*-
"""Task6.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/17MEIf7ATHw-2VxJZtA7QbSAAGqEOjf78?usp=sharing
"""

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# importing the data frame
data_frame = pd.read_csv('217005364-T0.csv')

# drop the unused columns
data_frame = data_frame.drop(['Date', 'High', 'Low', 'Close'], axis=1)

# computing class value for all records
for index in range(len(data_frame) - 1):
  curr_company = data_frame.iloc[index]['Company']
  curr_adj_close = data_frame.iloc[index]['Adj Close']

  next_company = data_frame.iloc[index + 1]['Company']
  next_adj_close = data_frame.iloc[index + 1]['Adj Close']

  if curr_company == next_company:
    if next_adj_close > curr_adj_close:
      data_frame.at[index, 'Direction'] = 1
    else :
      data_frame.at[index, 'Direction'] = 0
  else:
    data_frame.at[index, 'Direction'] = data_frame.iloc[index - 1]['Direction']

data_frame.at[len(data_frame) - 1, 'Direction'] = data_frame.iloc[len(data_frame) - 2]['Direction']

# assigning the features to X
X = data_frame[['Open', 'Adj Close', 'Volume', 'Company']]

# assigning the class variable to y
y = data_frame['Direction']

# extracting training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# initializing and training the classifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)

# testing classifier accuracy
y_predicted = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_predicted) * 100

print("Classification model is {:.2f}% acccurate".format(accuracy))
```
